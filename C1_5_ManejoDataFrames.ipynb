## Series
A Pandas *Series* is a one-dimensional array of indexed data.

A series can be created using various inputs like: Array, Dict, Scalar value or constant

```python
import pandas as pd
print(pd.__version__)
```

```python
data = pd.Series([0.25, 0.5, 0.75, 1.0])
data
type(data)
```

```python
data.index
data.values
```

```python
# modificando indices

data = pd.Series([0.25, 0.5, 0.75, 1.0],
                 index=['a', 'b', 'c', 'd'])
data
```

```python
population = pd.Series({'California': 38332521, 'Texas': 26448193, 'New York': 19651127,
                   'Florida': 19552860, 'Illinois': 12882135})
population
```

## DataFrame
is an analog of a two-dimensional array with both flexible row indices and flexible column names.

A pandas DataFrame can be created using various inputs like: Lists, dict, Series, Numpy ndarrays, Another DataFrame

```python
data = [1,2,3,4,5]
data ; type(data)

df = pd.DataFrame(data)
df ; type(df)
```

```python
df.index
df.columns
```

```python
df.ndim
df.shape
df.size
```

```python
# Crear la variable IE = INGRESOS/EDAD

clientes2 = clientes.copy()

clientes2['IE'] = clientes['INGRESOS'] / clientes['EDAD']
clientes2.head()
```

```python
# Crear un subconjunto de datos, donde el INGRESO es menor a 40000

clientes.sort_values('INGRESOS', ascending = False).head()

clientesS1 = clientes[clientes['INGRESOS'] < 40000]

clientesS1.sort_values('INGRESOS', ascending = False).head()
```

```python
# Crear un subconj. de datos, considerando solamente las mujeres

clientesS2 = clientes[clientes['SEXO'] == 'FEMENINO']

clientesS2.sort_values('SEXO', ascending = False).head()
```

```python
# Crear un subconj. de datos considerando clientes mayores a 35 aos, con hijos y viven en Lima este

clientesS3 = clientes[((clientes['EDAD'] >= 35) & (clientes['HIJOS'] > 0)) & (clientes['REGION'] == 'LIMA_ESTE')]

clientesS3.sort_values(['EDAD','HIJOS','REGION']).head()
```

## Combinacion de datasets

```python
band = pd.read_csv('band.csv')
instrum = pd.read_csv('instrument.csv')
band ; instrum
```

```python
# LEFT JOIN

pd.merge(band, instrum, how = 'left', on = 'name')
```

```python
# RIGHT JOIN

band ; instrum
pd.merge(band, instrum, how = 'right', on = 'name')
```

```python
# INNER JOIN

band ; instrum
pd.merge(band, instrum, how = 'inner', on = 'name')
```

```python
# OUTER JOIN

band ; instrum
pd.merge(band, instrum, how = 'outer', on = 'name')
```
