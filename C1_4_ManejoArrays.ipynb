# 4. Creacion de arrays

NumPy es un paquete de Python. Significa *Python Numérico*. Es una biblioteca que consta de objetos de matriz multidimensionales y una colección de rutinas para el procesamiento de matriz.

**Operaciones usando NumPy:**

* Operaciones matemáticas y lógicas en matrices.
* Transformaciones de Fourier y rutinas para la manipulación de formas.
* Operaciones relacionadas con el álgebra lineal. NumPy tiene funciones integradas para álgebra lineal y generación de números aleatorios.

```python
a1 + a2
a1 - a2
a1 * a2
a1 / a2
```

```python
np.exp(a1)
```

```python
np.log(a2)
```

```python
np.sqrt(a2)
```

```python
a3 + a4
a3 - a4
a3 * a4
a3 / a4
```

## Concatenacion de arrays

```python
x = np.array([1, 2, 3])
y = np.array([3, 2, 1])
x ; y

np.concatenate([x, y])
```

```python
grid = np.array([[1, 2, 3],
                 [4, 5, 6]])
grid
```

```python
# concatenate along the first axis
np.concatenate([grid, grid])
```

```python
# concatenate along the second axis (zero-indexed)
np.concatenate([grid, grid], axis=1)
```

```python
# Stack arrays

x = np.array([1, 2, 3])
x
y = np.array([[9, 8, 7], [6, 5, 4]])
y

# vertically stack the arrays
np.vstack([x, y])
```

```python
# horizontally stack the arrays
np.hstack([y[0,:], x])
```

```python
h = np.array([[99],
              [99]])
np.hstack([y, h])
```

## Arrays booleanos

```python
x = np.array([1, 2, 3, 4, 5])
x
```

```python
x == 3
x != 3
x < 3
x <= 3
x > 3
x >= 3
```

```python
b = x == 3
b.dtype
```

```python
np.random.seed(2021)
a = np.random.randint(10, size=(3, 4)) 
a
```

```python
a < 6
```

```python
# how many values less than 6?

np.count_nonzero(a < 6)
np.sum(a < 6)
```

```python
# how many values less than 6 in each ...

np.sum(a < 6, axis=1) # fila
np.sum(a < 6, axis=0) # columna
```

```python
# are there any values greater than 8?

np.any(a > 8)
np.any(a > 8, 1) # fila
np.any(a > 8, 0) # columna
```

```python
# are all values less than 10?

np.all(a < 10)
np.all(a < 10, 1) # fila
np.all(a < 10, 0) # columna
```

## Ordenamiento de Arrays

```python
x = np.array([2, 1, 4, 3, 5])
x
np.sort(x)
```

```python
# argsort: returns the indices of the sorted elements:

x = np.array([2, 1, 4, 3, 5])
x

i = np.argsort(x)
i

x[i]
```

```python
np.random.seed(2021)
a = np.random.randint(15, size=(3, 4)) 
a
```

```python
np.sort(a, 0) # ordenar por columna
np.sort(a, 1) # ordenar por fila
```
