```python
introduction = 'My name is {first_name} {middle_name} {last_name} AKA the {aka}.'
full_name = {
    'first_name': 'Tony',
    'middle_name': 'Howard',
    'last_name': 'Stark',
    'aka': 'Iron Man',
}
 
# Notice the use of "**" operator to unpack the values.
print(introduction.format(**full_name))
```

```python
 # %i signed decimal integer and %d signed decimal integer(base-10)

match = 12000
 
site = 'amazon'
 
print("%s is so useful. I tried to look\
up mobile and they had a nice one that cost %d rupees." % (site, match)) # %s => site ; %d => match
```
